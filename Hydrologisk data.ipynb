```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
```

```python
#BEREGNING AV JUSTERT TILSIG
# Lese CSV-fil
data = pd.read_csv('Vannføring korrigert.csv', sep=';')


kolonne_2 = data.iloc[:, 1]  # Henter alle rader fra kolonne 2
Q_s_m3 = kolonne_2.to_numpy()     #m^3/s    Konverter til numpy-array
Q_s = Q_s_m3 * 1000    #Tilsig Spannbogvatn [L/s]
datoer = pd.date_range(start='2009-12-01', periods=len(Q_s), freq='D')


A_s =  7.7    #Feltareal Spannbogvatn [km^2]
A_d =  1.4    #Feltareal Djupåvatn [km^2]
A_b =  2.9    #Feltareal Buvatn [km^2]

F_s = 66.6   #Årlig middelavrenning Spannbogvatn [L/s*km^2]
F_d = 73.7   #Årlig middelavrenning Djupåvatn [L/s*km^2]
F_b = 73     #Årlig middelavrenning Buvatn [L/s*km^2]

Q_d = Q_s * ((A_d*F_d)/(A_s*F_s))   #Justert tilsig Djupåvatn [L/s]
Q_b = Q_s * ((A_b*F_b)/(A_s*F_s))   #Justert tilsig Buvatn [L/s]


#Plotter tilsiget til Djupåvatn
plt.figure(figsize=(13, 6))
plt.plot(datoer,Q_d, color = 'blue', linewidth=0.5)
plt.xlabel('Tid')
plt.ylabel('Tilsig, [L/s]')
plt.title('Tilsig Djupåvatnet')
plt.xticks(rotation=60)
plt.xlim(datoer[1], datoer[-1])
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(bymonth=[1, 7]))  # Viser én dato per måned i Januar og Juli
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format: "Jan 2020", "Feb 2020"
plt.grid(axis='y', linestyle='--', alpha=0.5)
#plt.legend()
plt.show()


#Plotter tilsiget til Buvatn
plt.figure(figsize=(13, 6))
plt.plot(datoer,Q_b, color = 'green', linewidth=0.5)
plt.xlabel('Tid')
plt.ylabel('Tilsig, [L/s]')
plt.title('Tilsig Buvatnet')
plt.xticks(rotation=60)
plt.xlim(datoer[1], datoer[-1])
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(bymonth=[1, 7]))  # Viser én dato per måned i Januar og Juli
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format: "Jan 2020", "Feb 2020"
plt.grid(axis='y', linestyle='--', alpha=0.5)
#plt.legend()
plt.show()


#Plotter tilsiget til Spannbogvatn
plt.figure(figsize=(13, 6))
plt.plot(datoer,Q_s_m3, color = 'orange', linewidth=0.5)
plt.xlabel('Tid')
plt.ylabel(r'Tilsig, [$\mathrm{m}^3$/s]')
plt.title('Tilsig Spannbogvatn')
plt.xticks(rotation=60)
plt.xlim(datoer[1], datoer[-1])
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(bymonth=[1, 7]))  # Viser én dato per måned i Januar og Juli
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # Format: "Jan 2020", "Feb 2020"
plt.grid(axis='y', linestyle='--', alpha=0.5)
#plt.legend()
plt.show()




```

```python
#BEREGNER VARIGHETSKURVE TOTALT
# Beregn varighetskurve
def varighetskurve(tilsig):
    sortertQ = np.sort(tilsig)[::-1]  # Sorter i synkende rekkefølge
    varighet = np.arange(1., len(sortertQ) + 1) / len(sortertQ)
    return varighet, sortertQ

# Beregn varighetskurver for de justerte tilsigene
varighet_d, sortertQ_d = varighetskurve(Q_d)
varighet_b, sortertQ_b = varighetskurve(Q_b)


# Konverter y-aksen til prosent av middelvannføring
sortertQ_prosent_d = (sortertQ_d / Qm_d) * 100
sortertQ_prosent_b = (sortertQ_b / Qm_b) * 100

# Plot varighetskurver med x- og y-akse i prosent
plt.figure(figsize=(10, 6))
plt.plot(varighet_d * 100, sortertQ_prosent_d, label='Djupåvatn', color = 'blue')
#plt.plot(varighet_b * 100, sortertQ_prosent_b, label='Buvatn')
plt.xlabel('Varighet, [%]')
plt.ylabel('Vannføring i forhold til middelvannsføring [%]')
plt.title('Varighetskurve')
#plt.legend()
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

```python
#5-PERSENTIL
#Tar utgangspunkt i varighetskurvene for normalt år sommer og vinter (2014)

#indekser som angir rundt 95%:
#print(varighet_2014s)
#print(varighet_2014v)
indeks1 = -8
indeks2 = -9
indeks3 = -11
indeks4 = -12

#Varighet ved indekser: 
varighet_5p_2014s1 = varighet_2014s[indeks1]
varighet_5p_2014s2 = varighet_2014s[indeks2]
varighet_5p_2014v1 = varighet_2014s[indeks3]
varighet_5p_2014v2 = varighet_2014s[indeks4]

#Vannføring ved indekser: sortertQ_prosent_2014x[indeks] * Qm_vann_2014x
#Djupåvatnet
Q_5p_d_2014s1 = (sortertQ_prosent_2014s[indeks1]/100) * Qm_d_2014s   
Q_5p_d_2014s2 = (sortertQ_prosent_2014s[indeks2]/100) * Qm_d_2014s 
Q_5p_d_2014v1 = (sortertQ_prosent_2014v[indeks3]/100) * Qm_d_2014v 
Q_5p_d_2014v2 = (sortertQ_prosent_2014v[indeks4]/100) * Qm_d_2014v 

#Buvatnet
Q_5p_b_2014s1 = (sortertQ_prosent_2014s[indeks1]/100) * Qm_b_2014s 
Q_5p_b_2014s2 = (sortertQ_prosent_2014s[indeks2]/100) * Qm_b_2014s 
Q_5p_b_2014v1 = (sortertQ_prosent_2014v[indeks3]/100) * Qm_b_2014v
Q_5p_b_2014v2 = (sortertQ_prosent_2014v[indeks4]/100) * Qm_b_2014v


#Finner verdi for akkurat 95% ved å benytte lineær interpolasjon: verdi = x1 + ((x2 - x1)/(y2-y1) * (0.95 - y1)
#Djupåvatnet
Q_5p_d_2014s = Q_5p_d_2014s1 + (((Q_5p_d_2014s2 - Q_5p_d_2014s1)/(varighet_5p_2014s2 - varighet_5p_2014s1)) * (0.95 - varighet_5p_2014s1))
Q_5p_d_2014v = Q_5p_d_2014v1 + (((Q_5p_d_2014v2 - Q_5p_d_2014v1)/(varighet_5p_2014v2 - varighet_5p_2014v1)) * (0.95 - varighet_5p_2014v1))

#Buvatnet
Q_5p_b_2014s = Q_5p_b_2014s1 + (((Q_5p_b_2014s2 - Q_5p_b_2014s1)/(varighet_5p_2014s2 - varighet_5p_2014s1)) * (0.95 - varighet_5p_2014s1))
Q_5p_b_2014v = Q_5p_b_2014v1 + (((Q_5p_b_2014v2 - Q_5p_b_2014v1)/(varighet_5p_2014v2 - varighet_5p_2014v1)) * (0.95 - varighet_5p_2014v1))

print(f'5-persentil for Djupåvatnet sommeren et normalår (2014) er {Q_5p_d_2014s} L/s.')
print(f'5-persentil for Djupåvatnet vinteren et normalår (2014) er {Q_5p_d_2014v} L/s.')
print(f'5-persentil for Buvatnet sommeren et normalår (2014) er {Q_5p_b_2014s} L/s.')
print(f'5-persentil for Buvatnet vinteren et normalår (2014) er {Q_5p_b_2014v} L/s.')

#Varighetskurve sommer 2014
plt.figure(figsize=(11, 8.1))
plt.plot(varighet_2014s * 100, sortertQ_prosent_2014s, color = 'blue', lw=3)
plt.axhline(y=(Q_5p_b_2014s*100/Qm_b_2014s) , color='green', linestyle='--', label="5-persentil", lw=3)
plt.axvline(x=95, color='gray', linestyle='--', label="95% varighet", lw=3)
plt.xlabel('Varighet, [%]',fontsize = 20)
plt.ylabel('Vannføring i forhold til middelvannsføring, [%]',fontsize = 20)
#plt.title('Varighetskurve for sommer normalt år (2014): 01.05-30.09',fontsize = 15)
plt.tick_params(axis='both', labelsize=20)
plt.legend(loc='upper right',fontsize = 20)
plt.grid(linestyle='-', alpha=1)
plt.show()

#Varighetskurve vinter 2014
plt.figure(figsize=(11, 8.1))
plt.plot(varighet_2014v * 100, sortertQ_prosent_2014v, color = 'blue', lw=3)
plt.axhline(y=(Q_5p_b_2014v*100/Qm_b_2014v) , color='green', linestyle='--', label="5-persentil", lw=3)
plt.axvline(x=95, color='gray', linestyle='--', label="95% varighet", lw=3)
plt.xlabel('Varighet, [%]',fontsize = 20)
plt.ylabel('Vannføring i forhold til middelvannsføring, [%]',fontsize = 20)
plt.tick_params(axis='both', labelsize=20)
#plt.title('Varighetskurve for vinter normalt år (2014): 01.10-30.04',fontsize = 15)
plt.legend(loc='upper right',fontsize = 20)
plt.grid(linestyle='-', alpha=1)
plt.show()
```
